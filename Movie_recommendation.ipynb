```
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```
#load the dataset
movies = pd.read_csv("/content/Movies (1).tsv",sep="\t")
ratings = pd.read_csv("/content/Ratings (1).tsv",sep="\t")
```

```
#creating a mapping between id and title
idandtitle = {int(row["movieId"]): row["title"] for _,row in movies.iterrows()}
titleandid = {j:i for i,j in idandtitle.items()}
```

```
highratings = ratings[ratings.rating >=4]
itemsets = [
            [idandtitle[mov] for mov in highratings[highratings.userId == user].movieId]
            for user in highratings.userId
]

```

```
t = TransactionEncoder()
t_ary = t.fit(itemsets).transform(itemsets)
df = pd.DataFrame(t_ary, columns=t.columns_)
```

```
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True, max_len=2)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2)
df = pd.DataFrame(rules)
print(df[['antecedents','consequents']])
```
